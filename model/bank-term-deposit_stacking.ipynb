```python
# todo scatter plot

def remove_balance_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df4.groupby(['ageBin', 'job']):
        m = np.mean(subdf.balance)
        std = np.std(subdf.balance)
        reduced_df = subdf[(subdf.balance > (m-(std))) & (subdf.balance <= (m+(std)))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    
    return df_out

print(df4.shape)
df5 = remove_balance_outliers(df4)  

df5 = df5.dropna()
df5.shape
```

# Preprocessing

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import category_encoders as ce

categorical_columns = df5.select_dtypes(['object', 'bool', 'category']).columns.to_list()
numerical_columns = df5.select_dtypes('int64').columns.to_list()
numerical_columns = list(set(numerical_columns)-set(['y']))

high_card_cols = [c for c in categorical_columns if df5[c].nunique(dropna=True) >= 5]
low_card_cols  = [c for c in categorical_columns if df5[c].nunique(dropna=True) < 5]

low_card_cat_transformer = Pipeline(steps=[
    ('oneHot', OneHotEncoder(handle_unknown='ignore'))
])

high_card_cat_transformer = Pipeline([
    ("target_enc", ce.TargetEncoder(
        cols=high_card_cols,
        smoothing=10.0,
        min_samples_leaf=50
    ))
])

num_transformer = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer(transformers=[
    ("lowc", low_card_cat_transformer, low_card_cols),
    ("highc", high_card_cat_transformer, high_card_cols),
    ('num', num_transformer, numerical_columns)
])


X = df5[numerical_columns + categorical_columns]
y = df5['y']
```

```python
# # stacked model
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, StratifiedKFold
import lightgbm as lgb
from scipy.stats import randint, uniform, loguniform
from xgboost import XGBClassifier
from sklearn.svm import LinearSVC
from sklearn.base import BaseEstimator, ClassifierMixin


train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2, random_state=21)


class LGBMWithES(lgb.LGBMClassifier):
    def fit(self, X, y, **kwargs):
        X_tr, X_val, y_tr, y_val = train_test_split(
            X, y, test_size=0.2, stratify=y, random_state=21
        )
        return super().fit(
            X_tr, y_tr,
            eval_set=[(X_val, y_val)],
            eval_metric="auc",
            callbacks=[lgb.early_stopping(500)],
            # verbose=False,
            **kwargs
        )

class XGBWithES(XGBClassifier):
    def fit(self, X, y, **kwargs):
        X_tr, X_val, y_tr, y_val = train_test_split(
            X, y, test_size=0.2, stratify=y, random_state=21
        )
        return super().fit(
            X_tr, y_tr,
            eval_set=[(X_val, y_val)],
            **kwargs
        )

model_dict = {
    'lgbm': {
        'model': LGBMWithES(n_estimators=20000),
        'params': {
            # 'model__n_estimators': randint(350, 450),       # number of trees
            'model__max_depth': randint(6,9),            # tree depth
            'model__learning_rate': uniform(0.1, 0.2),   
            'model__num_leaves': randint(15, 63),
            'model__min_child_samples': randint(20, 100),
            'model__feature_fraction': uniform(0.6, 0.4),
            'model__bagging_fraction': uniform(0.6, 0.4),
            'model__bagging_freq': randint(1, 5),
            'model__lambda_l1': uniform(0, 1),
            'model__lambda_l2': uniform(1, 10), 
        }
    },
    'XGB': {
        'model': XGBWithES(
            tree_method="hist", 
            n_estimators=20000, 
            eval_metric="auc",
            verbose=False,
            early_stopping_rounds=500),
        'params': {
            # 'model__n_estimators': randint(350, 450),       # number of trees
            'model__max_depth': randint(8,10),            # tree depth
            'model__learning_rate': uniform(0.05, 0.08),
            'model__min_child_weight': randint(5, 20),
            'model__subsample': uniform(0.6, 0.4),           # 0.6–1.0
            'model__colsample_bytree': uniform(0.6, 0.4),    # 0.6–1.0
            'model__reg_alpha': uniform(0, 1),               # 0–1
            'model__reg_lambda': uniform(1, 10),     # shrinkage rate
        }
    }
}


cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=21)

def model_tuning(preprocessor, train_X, val_X, train_y, val_y, model_dict,cv):
    estimators = []
    for key, val in model_dict.items():
        model_abr = key
        model_type = val['model']
        params = val['params']


        bank_pipeline = Pipeline(steps=[
            ('preprocess', preprocessor),
            ('model', model_type)
        ])

        grid = RandomizedSearchCV(
            bank_pipeline,
            params,
            scoring='roc_auc',
            n_iter=30,
            cv=cv,
            n_jobs=-1,
            verbose=1,
        )

        grid.fit(train_X, train_y)
        best_params = grid.best_params_
        model_params = {key.replace('model__', ''): val for key, val in best_params.items()}

        base_params = model_type.get_params()
        base_params.update(model_params)
        tuned_model = type(model_type)(**base_params)

        estimators.append((model_abr, tuned_model))

        print(grid.best_score_)           # mean CV AUC across folds
        print(grid.cv_results_['mean_test_score'])
    
    return estimators

estimators = model_tuning(preprocessor, train_X, val_X, train_y, val_y, model_dict, cv)
```
